```python
import matplotlib.pyplot as plot    # we have to load matplotlib module so that we can plot our graph based on time series.
import pandas as pd                 # we will use panda module to extract our data or file, in this case we have file.
```

```python
file_path = r'C:\\Users\\abhij\\Desktop\\project\\data_csv.csv'    # In this case we have just given read permission for file.
data = pd.read_csv(file_path, index_col = 0)            # Panda module has been used to extract the data from the file.
```

```python
real_price = data['Real Price']            # real_price stores the data of 'Real Price' with help of data function
real_price.shape
real_earning = data['Real Earnings']       # real_earning stores the data of 'Real Earnings' with help of data function
Earnings = data['Earnings']
real_earning.shape
```

```python
%matplotlib notebook     
plot.plot(real_price, '.--r', label='Real Price')
plot.plot(real_price, '.--g', label='Real Price')
for i in range(len(real_price) - 1): # we are defining for loop so that we can change colour according to the price
    
    color = 'g' if real_price[i + 1] > real_price[i] else 'r'           # Green if price goes up, red if it goes down
    plot.plot([i, i+1], [real_price[i], real_price[i+1]], '.--'+color)

plot.plot(real_earning, '.--b', label='Real Earnings')                  # Plots graph regarding to 'Real Earnings' Data.

plot.xlabel('Date')             # we will label x axis as 'Date'
plot.ylabel('Price')            # we will label y axis as 'Price'
plot.title('Price Action')      # we will name the title as 'Price Action'
plot.grid()                     # we will use .grid function to visually aligning data points on the plot
plot.legend()                   # we will use .legend function which describes the data from the set.
```
